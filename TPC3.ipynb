## TPC3: Adivinha o número!?

Desenvolve em Python o código necessário para o jogo do Adivinha o número: 
* O jogo deverá ter dois modos: o computador tenta acertar no número que o jogador pensou ou, o jogador tenta acertar no número que o computador pensou; 
* No fim, o programa deverá indicar quantas tentativas foram realizadas para chegar ao resultado;
* O Jogo: 
    - o utilizador pensa num número entre 0 e 100;
    - o computador tenta adivinhar indicando um número do intervalo;
    - o utilizador indica se o número em que pensou é **maior**, **menor**, ou se o computador **acertou**;
    - ao terminar, o programa imprime o número de tentativas que o computador teve de fazer até chegar ao número.

Que conclusão podes tirar sobre o número de tentativas necessárias para um número entre 0 e 100? E se fosse entre 0 e 1000?

```python
from random import randint



def computador_adivinha():
    n_pos=range(0,101)
    tentativas=0
    print("Pense num número inteiro entre 0 e 100")

    while True:
        n=round(len(n_pos)/2)
        print("O número é:"+ str(n_pos[n]))
        variavel=input("O número é:"+ str(n_pos[n]) + "? Responda 'maior', 'menor' ou 'certo'?").lower()

        if variavel == 'maior':
            n_pos=n_pos[n:]
            tentativas = tentativas + 1

        elif variavel == 'menor':
            n_pos=n_pos[:n]
            tentativas = tentativas + 1
        
        elif variavel == 'certo':
            print("O seu número foi encontrado. É o " + str(n_pos[n]) + "E demorou" + str(tentativas) + "tentativas!")
            break

        else:
            print("Deve inserir: 'maior', 'menor' ou 'certo'.")

    
    def jogador_adivinha():
        n=randint(0, 100)
        tentativas=0

        while True:
            tentativas = tentativas + 1
            variavel = int(input("Adivinhe um número entre 0 e 100:"))

            if variavel < n:
                print ("Menor!")

            elif variavel > n:
                print ("Menor!")

            else:
                print("Parabéns! Adivinhou o número em " + str(tentativas) + "tentativas.")
                break

def selecionar_modo_jogo():

        while True:
            modo=input("Escolha o modo de jogo que pretende (digite 'jogador' ou 'computador'):")

            if modo == 'jogador':
                jogador_adivinha()
                break

            elif modo == 'computador':
                computador_adivinha()
                break

            else:
                print ("Modo de jogo inválido, digite 'jogador' ou 'computador'!")

        selecionar_modo_jogo()

    


```
